```python
key = lookup[['State', 'Abbreviation']]
```

```python
rawdata.merge(key, left_on='State', right_on='Abbreviation')
```

```python
s = lookup.set_index('State')['Abbreviation']
s.index = s.index.str.strip(' ')
s.head()
```

```python
m = rawdata['State'].map(s)
m.to_frame

incidents = m.value_counts(dropna=False)
```

```python
incidents = pd.DataFrame(incidents)
incidents.rename(index=str, columns={ 0 : 'Incidents'})
incidents_index = pd.DataFrame(incidents.index)
type(incidents_index)
incidents_index
```

```python
ndf = pd.DataFrame()
ndf['State'] = incidents_index
ndf['Incidents'] = incidents

ndf

>>> d = {'col1': [1, 2], 'col2': [3, 4]}
>>> df = pd.DataFrame(data=d)

d = { 'State': incidents_index, 'Incidents' }
```

```python
# viridis
brew = ["#ffffd9","#edf8b1","#c7e9b4","#7fcdbb","#41b6c4","#1d91c0","#225ea8","#253494","#081d58"]
# plotly colorscale format
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]
```

```python
data = [ dict(
        type='choropleth',
        colorscale = 'Viridis',
        autocolorscale = False,
        locations = incidents.index,
        z = incidents['State'].astype(int),
        locationmode = 'USA-states',
        # text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Number of Mass Shootings in the United States From July 23, 2016 to November 16, 2017")
        ) ]
```

```python
layout = dict(
        title = '',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
py.iplot( fig, filename='mass-shootings-incidents-choropleth-map' )
```
